```
import malaya
```

```
modelsub = malaya.subjective.multinomial()
modelsent = malaya.sentiment.multinomial()
```

```
import pandas as pd
```

```
df = pd.read_csv('/content/mxinpqtc.csv')
```

```
df
```

```
data = df['text']
```

```
data = [str(x) for x in data]
```

```
len(data)
```

```
type(data)
```

```
sub = modelsub.predict_proba(data, add_neutral = False)
sent = modelsent.predict_proba(data, add_neutral = False)
```

```
sub_score = []
for i in range(len(sub)):
    
    sub_score.append(sub[i]['positive'])

sent_score = []
for i in range(len(sent)):
    
    sent_score.append(sent[i]['negative'])

```

```
sub_sent_df = pd.DataFrame()
```

```
sub_sent_df['subjective'] = sub_score
```

```
sub_sent_df['sentiment'] = sent_score
```

```
df = pd.concat([df, sub_sent_df], axis=1)
```

```
df
```

```
top_100_sent = df.sort_values(by=('sentiment'), ascending=False).iloc[:100]
top_100_sent
```

```
list(top_100_sent['text'])
```
